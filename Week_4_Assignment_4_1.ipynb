```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset from the uploaded file
df = pd.read_csv('Titanic.csv')

# Display the first few rows to verify loading
df.head()
```

```python
sns.catplot(x="sex", y="survived", kind="bar", data=df)
plt.title("Survival Rate by Gender")
plt.ylabel("Survival Rate")
plt.show()
# Insight: The survival rate was significantly higher for females than males, indicating a possible priority given to women during the evacuation.
```

```python
sns.catplot(x="class", y="survived", kind="bar", data=df)
plt.title("Survival Rate by Passenger Class")
plt.ylabel("Survival Rate")
plt.show()
# Insight: First-class passengers had a higher survival rate compared to second- and third-class passengers, suggesting a class-based rescue priority.
```

```python
sns.boxplot(x="survived", y="age", data=df)
plt.title("Age Distribution of Survivors vs Non-Survivors")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Age")
plt.show()
# Insight: Survivors tend to have a slightly younger median age than non-survivors, indicating that younger passengers may have had a better chance of survival.
```

```python
sns.violinplot(x="survived", y="fare", data=df)
plt.title("Fare Distribution by Survival")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Fare")
plt.show()
# Insight: Survivors paid higher fares on average, with a wider range of fares, suggesting that wealthier passengers had better survival odds.
```
