```python
import pandas as pd
```

```python
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
```

```python
#Print the first 5 rows to observe your data
iris_df.head()
```

Task 1 ("1. Compute mean, median, and mode for each feature.").

```python
# 1. Compute mean, median, and mode for each feature
# Select only the numeric columns for statistical calculations
numeric_columns = iris_df.select_dtypes(include=['number'])

mean_values = numeric_columns.mean()
median_values = numeric_columns.median()
mode_values = numeric_columns.mode().iloc[0]

print("Mean Values:\n", mean_values)
print("\nMedian Values:\n", median_values)
print("\nMode Values:\n", mode_values)

# If you want to see the mode of the species column separately
species_mode = iris_df['species'].mode()[0]  # Assuming 'species' is the column name
print("\nMode of Species:", species_mode)
```

Task 2 ("2. Calculate the range, variance, and standard deviation for each feature.")

```python
# 2. Calculate the range, variance, and standard deviation for each feature
# Select only numeric columns for calculations
numeric_columns = iris_df.select_dtypes(include=['number']).columns
range_values = iris_df[numeric_columns].max() - iris_df[numeric_columns].min()
variance_values = iris_df[numeric_columns].var()
std_dev_values = iris_df[numeric_columns].std()

print("Range Values:\n", range_values)
print("\nVariance Values:\n", variance_values)
print("\nStandard Deviation Values:\n", std_dev_values)
```

Task 3 ("3. Compute the 25th, 50th, and 75th percentiles (quartiles) for each feature.")

```python
# 3. Compute the 25th, 50th, and 75th percentiles (quartiles) for each feature
# Select only numeric columns for quantile calculation
numeric_columns = iris_df.select_dtypes(include=['number'])
percentiles = numeric_columns.quantile([0.25, 0.50, 0.75])
print("Percentiles (25th, 50th, 75th):\n", percentiles)
```

Task 4 ("4. Create a histogram for each of the features to visualize their distributions.")

```python
# 4. Create a histogram for each of the features
import matplotlib.pyplot as plt
iris_df.hist(figsize=(10, 8))
plt.suptitle('Histograms of Iris Features')
plt.show()
```

Task 5 ("5. Plot scatter plots between pairs of features to explore relationships.")

```python
# 5. Plot scatter plots between pairs of features
import seaborn as sns
sns.pairplot(iris_df)
plt.suptitle('Scatter Plots of Iris Features')
plt.show()
```
